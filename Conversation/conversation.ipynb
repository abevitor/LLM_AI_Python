```python
from openai import OpenAI

OLLAMA_BASE_URL = "http://localhost:11434/v1"

client = OpenAI(
    base_url=OLLAMA_BASE_URL,
    api_key="ollama"
)

MODEL = "llama3.2"


conversation = ""

def talk(name, personality, conversation):
    system_prompt = f"""
You are {name}.
{personality}
You are in a conversation with other people.
"""

    user_prompt = f"""
Conversation so far:
{conversation}

Now respond as {name}.
"""

    response = client.chat.completions.create(
        model=MODEL,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}
        ]
    )

    reply = response.choices[0].message.content
    return reply



alex_personality = "You are very argumentative and disagree with everything in a sarcastic way."
blake_personality = "You are calm and logical."
charlie_personality = "You are funny and make jokes."

for i in range(5):
    
    alex_reply = talk("Alex", alex_personality, conversation)
    conversation += f"\nAlex: {alex_reply}"
    
    blake_reply = talk("Blake", blake_personality, conversation)
    conversation += f"\nBlake: {blake_reply}"
    
    charlie_reply = talk("Charlie", charlie_personality, conversation)
    conversation += f"\nCharlie: {charlie_reply}"


print(conversation)
```
