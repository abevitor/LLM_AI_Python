```python
from openai import OpenAI


OLLAMA_BASE_URL = "http://localhost:11434/v1"

client = OpenAI(
    base_url=OLLAMA_BASE_URL,
    api_key="ollama" #rodando local
)

MODEL_LLAMA = "llama3.2"

def explain_code(question):
    system_prompt = """
You are a helpful assistant that explains technical concepts clearly and simply.
Always break down the explanation step by step and give examples when useful.
"""

    response = client.chat.completions.create(
        model=MODEL_LLAMA,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": question}
        ],
        temperature=0.3 # de uma forma simples de entender, eh o "nive de criatividade"
    )

    return response.choices[0].message.content



question = """
Please explain what this code does and why:
yield from {book.get("author") for book in books if book.get("author")}
"""


answer = explain_code(question)

print(answer)
```
